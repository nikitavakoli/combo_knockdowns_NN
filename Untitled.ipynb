## Default PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Step 1: Read the CSV files
file_paths = ['data/KRAS_CAF.csv', 'data/KRAS_CRC.csv', 'data/WT_CAF.csv', 'data/WT_CRC.csv']
dfs = [pd.read_csv(file) for file in file_paths]

# Add a column to each DataFrame indicating the source file
for i, df in enumerate(dfs):
    df['Source'] = f'File {i+1}'

# Step 2: Combine the DataFrames (assuming they have the same columns)
combined_df = pd.concat(dfs, ignore_index=True)

# Step 3: Ensure only numeric data is used for PCA
numeric_df = combined_df.select_dtypes(include=[float, int])

# Handle missing values by imputing with the mean
imputer = SimpleImputer(strategy='mean')
numeric_df = imputer.fit_transform(numeric_df)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Step 4: Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])

# Add the source information back
principal_df['Source'] = combined_df['Source']

# Step 5: Visualize the Results with color-coding
plt.figure(figsize=(10, 10))

# Define colors for each source file
colors = {'File 1': 'r', 'File 2': 'g', 'File 3': 'b', 'File 4': 'y'}

# Plot with colors according to the source file
for source in principal_df['Source'].unique():
    subset = principal_df[principal_df['Source'] == source]
    plt.scatter(subset['Principal Component 1'], subset['Principal Component 2'], label=source, color=colors[source])

plt.title('PCA Projection into 2D Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Step 1: Read the CSV files
file_paths = ['data/KRAS_CAF.csv', 'data/KRAS_CRC.csv', 'data/WT_CAF.csv', 'data/WT_CRC.csv']
dfs = [pd.read_csv(file) for file in file_paths]

# Add a column to each DataFrame indicating the source file
for i, df in enumerate(dfs):
    df['Source'] = f'File {i+1}'

# Step 2: Combine the DataFrames (assuming they have the same columns)
combined_df = pd.concat(dfs, ignore_index=True)

# Step 3: Ensure only numeric data is used for PCA
numeric_df = combined_df.select_dtypes(include=[float, int])

# Handle missing values by imputing with the mean
imputer = SimpleImputer(strategy='mean')
numeric_df = imputer.fit_transform(numeric_df)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Step 4: Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Print explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by the principal components: {explained_variance}')

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])

# Add the source information back
principal_df['Source'] = combined_df['Source']

# Step 5: Visualize the Results with color-coding
plt.figure(figsize=(10, 10))

# Define colors for each source file
colors = {'File 1': 'r', 'File 2': 'g', 'File 3': 'b', 'File 4': 'y'}

# Plot with colors according to the source file
for source in principal_df['Source'].unique():
    subset = principal_df[principal_df['Source'] == source]
    plt.scatter(subset['Principal Component 1'], subset['Principal Component 2'], label=source, color=colors[source])

plt.title('PCA Projection into 2D Space')
plt.xlabel(f'Principal Component 1 ({explained_variance[0]:.2f} variance)')
plt.ylabel(f'Principal Component 2 ({explained_variance[1]:.2f} variance)')
plt.legend()
plt.grid()

plt.show()
```

## Default TSNE

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Step 1: Read the CSV files
file_paths = ['data/KRAS_CAF.csv', 'data/KRAS_CRC.csv', 'data/WT_CAF.csv', 'data/WT_CRC.csv']
dfs = [pd.read_csv(file) for file in file_paths]

# Add a column to each DataFrame indicating the source file
for i, df in enumerate(dfs):
    df['Source'] = f'File {i+1}'

# Step 2: Combine the DataFrames (assuming they have the same columns)
combined_df = pd.concat(dfs, ignore_index=True)

# Step 3: Ensure only numeric data is used for t-SNE
numeric_df = combined_df.select_dtypes(include=[float, int])

# Handle missing values by imputing with the mean
imputer = SimpleImputer(strategy='mean')
numeric_df = imputer.fit_transform(numeric_df)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Step 4: Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(scaled_data)

# Create a DataFrame with the t-SNE results
tsne_df = pd.DataFrame(data=tsne_results, columns=['TSNE Component 1', 'TSNE Component 2'])

# Add the source information back
tsne_df['Source'] = combined_df['Source']

# Step 5: Visualize the Results with color-coding
plt.figure(figsize=(10, 10))

# Define colors for each source file
colors = {'File 1': 'r', 'File 2': 'g', 'File 3': 'b', 'File 4': 'y'}

# Plot with colors according to the source file
for source in tsne_df['Source'].unique():
    subset = tsne_df[tsne_df['Source'] == source]
    plt.scatter(subset['TSNE Component 1'], subset['TSNE Component 2'], label=source, color=colors[source])

plt.title('t-SNE Projection into 2D Space')
plt.xlabel('TSNE Component 1')
plt.ylabel('TSNE Component 2')
plt.legend()
plt.grid()

plt.show()
```
